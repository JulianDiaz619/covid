```python
import pandas as pd
```

```python
json = "https://data.cdc.gov/resource/n8mc-b4w4.json?$limit=10000000"
df = pd.read_json(json)
df.head(10)
```

```python
df = df.drop(['state_fips_code', 'county_fips_code', 'ethnicity', 'process', 'exposure_yn', 'case_onset_interval', 'case_positive_specimen', 'exposure_yn', 'case_onset_interval', 'case_positive_specimen'], axis=1)
df.head(10)
```

```python
df["case_month"] = pd.to_datetime(df["case_month"], format='%Y-%m')
df.dtypes
```

```python
len(df)
```

```python
# What rows needed to be dropped through dropna()
print(f"Age Group column contains: {df['age_group'].unique()}")
print(f"Sex column contains: {df['sex'].unique()}")
print(f"Race column contains: {df['race'].unique()}")
```

```python
df.head(10)
```

```python
age_df = df.copy()
age_df = age_df.drop(['sex', 'race'], axis=1)
age_df.head()
```

```python
sex_df = df.copy()
sex_df = sex_df.drop(['age_group', 'race'], axis=1)
sex_df.head()
```

```python
race_df = df.copy()
race_df = race_df.drop(['age_group', 'sex'], axis=1)
race_df.head()
```

```python
age_df['age_group'].replace('Missing', pd.NA, inplace=True)
age_df.dropna(subset=['age_group'], inplace = True)
age_df = age_df[age_df['age_group'] != 'NA']
age_df.head(20)
```

```python
print(f"Age Group column contains: {age_df['age_group'].unique()}")
print(f"Length of age_df: {len(age_df)}")
```

```python
sex_df['sex'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
sex_df = sex_df[sex_df['sex'] != 'NA']
sex_df.dropna(subset=['sex'], inplace = True)
sex_df.head(20)
```

```python
print(f"Sex column contains: {sex_df['sex'].unique()}")
print(f"Length of sex_df: {len(sex_df)}")
```

```python
race_df['race'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
race_df = race_df[race_df['race'] != 'NA']
race_df.dropna(subset=['race'], inplace = True)
race_df.head(20)
```

```python
print(f"Race column contains: {race_df['race'].unique()}")
print(f"Length of race_df: {len(race_df)}")
```

```python
age_df.to_csv('resources/demographic_analysis/age.csv', index = False)
```

```python
sex_df.to_csv('resources/demographic_analysis/sex.csv', index = False)
```

```python
race_df.to_csv('resources/demographic_analysis/race.csv', index = False)
```
