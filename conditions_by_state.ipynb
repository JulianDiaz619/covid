```python
states = ['United States', 'Montana', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'New York City', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Nebraska', 'Missouri', 'Alabama', 'Mississippi', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Wyoming', 'District of Columbia']

state_data = []

for state in states:
    state_df = df[df['State'] == state]
    if not state_df.empty:
        state_grouped = state_df.groupby(['State', 'Condition Group', 'Condition']).sum().reset_index()
        state_data.append(state_grouped)
```

```python
import os

# Set the path to the output directory
output_dir = os.path.join(os.path.expanduser('~'), 'Desktop', 'final_project')

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through the state names and export the corresponding DataFrame to a CSV file
for state, state_df in state_dfs.items():
    # Set the output path for the current state's CSV file
    output_path = os.path.join(output_dir, state + '.csv')
    # Export the current state's DataFrame to a CSV file
    state_df.to_csv(output_path, index=False)
```
