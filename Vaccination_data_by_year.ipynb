```python
import pandas as pd
```

```python
covid_vacc_df = pd.read_csv('resources/us_state_vaccinations.csv')
covid_vacc_df.head()
```

```python
covid_vacc_df.isnull().sum()
```

```python
covid_vacc_df.dtypes
```

```python
covid_vacc_df_drop = covid_vacc_df.drop(columns=["total_vaccinations","total_distributed","total_vaccinations_per_hundred","distributed_per_hundred","daily_vaccinations_raw","daily_vaccinations","daily_vaccinations_per_million","share_doses_used","total_boosters","total_boosters_per_hundred"])
covid_vacc_df_drop.head()
```

```python
covid_vacc_df_drop.isnull().sum()
```

```python
covid_vacc_df_dropnull = covid_vacc_df_drop.dropna()
covid_vacc_df_dropnull.isnull().sum()
```

```python
covid_vacc_df_dropnull.head()
```

```python
covid_vacc_df_copy = covid_vacc_df_dropnull.copy()
```

```python
covid_vacc_df_copy['date'] = pd.to_datetime(covid_vacc_df_copy['date'])
covid_vacc_df_copy['year'] = covid_vacc_df_copy['date'].dt.year
```

```python
covid_vacc_df_copy.head()
```

```python
covid_vacc_df_1 = covid_vacc_df_copy.drop(columns=["date"])
covid_vacc_df_1.head()
```

```python
covid_vacc_df_2 = covid_vacc_df_1[["year","location","people_vaccinated","people_vaccinated_per_hundred","people_fully_vaccinated","people_fully_vaccinated_per_hundred"]]
covid_vacc_df_2.head()
```

```python
last_vaccinations_df = pd.DataFrame(columns=['location', 'people_vaccinated','people_fully_vaccinated','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred'])

for state in covid_vacc_df_2['location'].unique():
    for year in [2021, 2022, 2023]:
        last_vaccinations_df = last_vaccinations_df.append(
            covid_vacc_df_2.loc[(covid_vacc_df_2['location']==state) & (covid_vacc_df_2['year']==year)].tail(1)[['location', 'people_vaccinated','people_fully_vaccinated','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred', 'year']]
        )

last_vaccinations_df.reset_index(drop=True, inplace=True)
last_vaccinations_df.head()
```

```python
grouped_df = last_vaccinations_df.groupby(['location', 'year']).agg({'people_vaccinated': 'sum','people_fully_vaccinated':'sum','people_vaccinated_per_hundred':'sum','people_fully_vaccinated_per_hundred':'sum'})
grouped_df.head()
```

```python
grouped_df.to_csv('vaccinattion_data_unformat.csv', index=True)
```

```python
covid_vacc_year_df = pd.read_csv('/resources/vaccinattion_data_unformat.csv')
covid_vacc_year_df.head()
```

```python
covid_vacc_year_df.dtypes
```

```python
covid_vacc_year_df['location'] = pd.Categorical(covid_vacc_year_df['location'])
```

```python
covid_vacc_year_df['year'] = covid_vacc_year_df['year'].astype(int)
```

```python
covid_vacc_year_df.dtypes
```

```python
# Calculate the total population based on the vaccination data
covid_vacc_year_df['total_population'] = (covid_vacc_year_df['people_vaccinated'] / (covid_vacc_year_df['people_vaccinated_per_hundred'] / 100))
```

```python
covid_vacc_year_df.head()
```

```python
# Define list of US state names
us_states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware','Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky','Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi','Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico','New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania','Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont','Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']

# Filter data for US states
us_states_df = covid_vacc_year_df[covid_vacc_year_df['location'].isin(us_states)]

# Create new dataframe with selected columns
us_states_vacc_df = us_states_df[['location', 'year', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'total_population']]
```

```python
us_states_vacc_df.head()
```

```python
us_states_vacc_df.to_csv('vaccinattion_data_by_year.csv', index=False)
```
