# Cleaned DataFrame for Machine Learning

```python
clean_df = df.copy()
clean_df = clean_df.drop(['case_month', 'res_state', 'res_county', 'hosp_yn'], axis=1)
clean_df.head(10)
```

```python
for col in clean_df.columns: 
    print(col, df[col].unique())
```

```python
clean_df['age_group'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
clean_df['sex'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
clean_df['race'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
clean_df['symptom_status'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
clean_df['icu_yn'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)
clean_df['death_yn'].replace({'Unknown': pd.NA, 'Missing': pd.NA}, inplace=True)

clean_df = clean_df[clean_df[['age_group', 'sex', 'race', 'symptom_status', 'icu_yn', 'death_yn']] != 'NA']
clean_df.dropna(subset=['age_group', 'sex', 'race', 'symptom_status', 'icu_yn', 'death_yn'], inplace = True)
clean_df.head()
```

```python
for col in clean_df.columns: 
    print(col, clean_df[col].unique())
```

```python
new_clean_df = clean_df.drop(['current_status', 'underlying_conditions_yn'], axis=1)
new_clean_df['death_yn'] = new_clean_df['death_yn'].replace({'Yes': 1, 'No': 0})
new_clean_df.head()
```

```python
new_clean_df.dtypes
```

```python
encoded_df = pd.get_dummies(new_clean_df, columns = ['age_group', 'sex', 'race', 'symptom_status', 'icu_yn'])
encoded_df.head(25)
```

```python
encoded_df['death_yn'].unique()
```

```python
# Set up for machine learning mortality model
y = encoded_df['death_yn']
X = encoded_df.drop(columns= 'death_yn')
```

# Logistic Regression Attempt

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y, random_state=1, stratify=y)
```

```python
# Initiate similar to Module #18

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=200,
   random_state=1)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print("\nClassification Report: \n")
report = classification_report(y_test, y_pred)
print(report)
```

```python
cm_log = confusion_matrix(y_test, y_pred)
cm_log_df = pd.DataFrame(
    cm_log, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_log_df
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
X_train_RF, X_test_RF, y_train_RF, y_test_RF = train_test_split(X, y, random_state=1)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train_RF)

X_train_scaled_RF = X_scaler.transform(X_train_RF)
X_test_scaled_RF = X_scaler.transform(X_test_RF)
```

```python
rf_model = RandomForestClassifier(n_estimators=128, random_state=1) 
rf_model = rf_model.fit(X_train_scaled_RF, y_train_RF)
```

```python
predictions_RF = rf_model.predict(X_test_scaled_RF)
cm = confusion_matrix(y_test_RF, predictions_RF)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
```

```python
acc_score_RF = accuracy_score(y_test_RF, predictions_RF)
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score_RF}")
print("Classification Report")
print(classification_report(y_test_RF, predictions_RF))
```
